```python
def is_number_in_list(numbers, searched_number):
    for number in numbers:
        if number == searched_number:
            return True
    return False
```

```python
def is_number_in_list(numbers, searched_number):
    list_with_just_searched_numbers = [number
                                      for number in numbers
                                      if number == searched_number]
    return len(list_with_just_searched_numbers) > 0
```

```python
def how_many_of_each_item_are_there(a_list):
    counts = {}
    for item in a_list:
        previous_count_of_item_or_zero = counts.get(item, 0)
        counts[item] = previous_count_of_item_or_zero + 1
    return counts

# or even
def how_many_of_each_item_are_there(a_list):
    counts = {}
    for item in a_list:
        counts[item] = counts.get(item, 0) + 1
    return counts


# hint: counts.get("apple", 0)   will return the value in the key of "apple", or 0 if there is no value in that key
```

```python
def how_many_of_each_item_are_there(a_list):
    #     grab a list of unique keys in the input
    unique_items = set(a_list)

    #     create a list of counts for these keys
    key_counts = [a_list.count(item)
                 for item in unique_items]

    #     'zip' these two lists together with into a dict. 
    # zip works like this: zip(keys, values) -> dict with keys and values
    # example: dict(zip(["apple", "plum"], [2,4])) == {'apple': 2, 'plum': 4}

    counts = dict( zip(unique_items, key_counts) )  
    return counts
```

```python
def return_city_names_with_density_lower_than(city_list, max_density):
    just_names = []
    for city in city_list:
        density_of_this_city = city['population'] / city['area']
        print(f"density of {city['name']} is {density_of_this_city}") # for debugging
        
        if density_of_this_city < max_density:
            just_names.append(city['name'])
            
    print(f"cities with density below {max_density} are {just_names}") # for debugging

    return just_names
```

```python
def return_city_names_with_density_lower_than(city_list, max_density):
    just_names = [city['name']
                 for city in city_list
                 if city['population'] / city['area'] < max_density]
            
    print(f"cities with density below {max_density} are {just_names}") # for debugging

    return just_names
```

```python
cities1 = [  {"name":"Edinburgh",  "population":500000, "area":264},
                 {"name":"Glasgow",  "population":600000, "area":175},
                 {"name":"Inverness", "population":50000, "area":20}]

assert return_city_names_with_density_lower_than(cities1, 3500) == ["Edinburgh" ,"Glasgow", "Inverness"]
assert return_city_names_with_density_lower_than(cities1, 3400) ==  ["Edinburgh" , "Inverness"]
assert return_city_names_with_density_lower_than(cities1, 2000) ==  ["Edinburgh" ]
assert return_city_names_with_density_lower_than(cities1, 1000) ==  [ ]
assert return_city_names_with_density_lower_than([], 3500) == []
print("all tests passed")
```

Turn inromation about opening hours of a shop into a string and print it. See example below, but feel free to change it. Use string manipulation and loops to produce the type of sign you could see on the door of a shop



```python
opening_hours = [{'day':'Monday', 'open': 9, 'close':18},
                {'day':'Tuesday', 'open': 12, 'close':18},
                {'day':'Wednesday'},
                {'day':'Thursday', 'open': 9, 'close':18},
                {'day':'Friday', 'open': 9, 'close':21},
                {'day':'Saturday', 'open': 12, 'close':18},
                {'day':'Sunday', 'open': 12, 'close':18}]

# what are you going to do with days when they are closed? Remember you can always check for None with
# if day_data.get('open') != None:
```

```python
def top_part():
    result_so_far = ""
    welcome_message = 'WELCOME'
    result_so_far += f"{welcome_message:*^30}\n"
    result_so_far += f"{'':*^30}\n"
    return result_so_far

def bottom_part():
    result_so_far = ""
    result_so_far += f"{'':*^30}\n"
    return result_so_far

def day_part(day_info):
    opening_hours_text = ""
    if day_info.get('open') == None or day_info.get('close') == None:
        opening_hours_text = 'Closed'
    else:
        opening_hours_text += f"{day_info.get('open'):2} till {day_info.get('close'):2}"
    return f"* {day_info['day']:.<10}{opening_hours_text:.^16} *\n"

def create_sign(week_data):
    final_sign_so_far = ""
    final_sign_so_far += top_part()
    for day_data in week_data:
        final_sign_so_far += day_part(day_data)
    final_sign_so_far += bottom_part()
    return final_sign_so_far
```
